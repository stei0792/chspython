# Up for crude and coal = add dollar to each month
# Down for crude and coal = subtract dollar to each month
# Up for corn = add 0.5 dollar to each month
# Down for corn = subtract 0.5 dollar to each month
## Model uses 1/5/1996 - 3/25/2016, same parameters from (Urea Model Using My Predictions)
## Only need 1 coal since the coal values (RDF,MLR) only changed in prediction section

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
data = pd.read_csv("C:/Users/msteinme/Documents/ureascenarios.csv")
#model part
df_96on = data[['Date','Crude_Same','Corn_Same','Coal_Same','Crude_Up','Crude_Down','Corn_Up','Corn_Down','Coal_Up','Coal_Down', 'Urea']][0:1056]
df_96on.tail()
```

```python
from datetime import datetime
df_96on['Date'] = pd.to_datetime(df_96on['Date'])
date = df_96on['Date']
df_96on.dtypes
```

```python
X = df_96on[['Crude_Same','Corn_Same','Coal_Same']].values #coal_rdf(rdf) same as the mlr one but not later on in predicted sec
y = df_96on['Urea'].values
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3,random_state=1)
forest=RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=10,
           max_features='auto', max_leaf_nodes=None, min_samples_leaf=1,
           min_samples_split=2, min_weight_fraction_leaf=0.0,
           n_estimators=200, n_jobs=1, oob_score=False, random_state=None,
           verbose=0, warm_start=False)
forest.fit(X_train, y_train)
y_train_pred= forest.predict(X_train)
y_test_pred= forest.predict(X_test)
print('MSE train: %.3f, test: %.3f' % (mean_squared_error(y_train, y_train_pred),mean_squared_error(y_test, y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (r2_score(y_train, y_train_pred),r2_score(y_test,y_test_pred)))
print(forest.feature_importances_)
```

```python
#this part of models prediction same for urea_pred_crudeup, crudedown, cornup, corndown, gasup, gasdown
#crude
urea_predicted3 = forest.predict(X)
urea_predicted3 = pd.DataFrame(urea_predicted3)
df_96on['Urea_Pred_CrudeUp_RDF'] = urea_predicted3
rdf_error3 = (abs(df_96on['Urea_Pred_CrudeUp_RDF'] - df_96on['Urea'])/df_96on['Urea'])*100
df_96on['RDF_Error'] = rdf_error3

urea_predicted4 = forest.predict(X)
urea_predicted4 = pd.DataFrame(urea_predicted4)
df_96on['Urea_Pred_CrudeDown_RDF'] = urea_predicted4

#corn
urea_predicted5 = forest.predict(X)
urea_predicted5 = pd.DataFrame(urea_predicted5)
df_96on['Urea_Pred_CornUp_RDF'] = urea_predicted5

urea_predicted6 = forest.predict(X)
urea_predicted6 = pd.DataFrame(urea_predicted6)
df_96on['Urea_Pred_CornDown_RDF'] = urea_predicted6

#coal
urea_predicted7 = forest.predict(X)
urea_predicted7 = pd.DataFrame(urea_predicted7)
df_96on['Urea_Pred_CoalUp_RDF'] = urea_predicted7

urea_predicted8 = forest.predict(X)
urea_predicted8 = pd.DataFrame(urea_predicted8)
df_96on['Urea_Pred_CoalDown_RDF'] = urea_predicted8

#same
urea_predicted9 = forest.predict(X)
urea_predicted9 = pd.DataFrame(urea_predicted9)
df_96on['Urea_Pred_Same_RDF'] = urea_predicted9

print("Average RDF error is: %.3f" % (rdf_error3.mean()))
df_96on
```
